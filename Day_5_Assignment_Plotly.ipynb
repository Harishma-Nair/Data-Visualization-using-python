1) What is Plotly?

Plotly is a collaborative, web-based graphing and analytics platform.  

2)Where is Plotly used ?

It allows us to import, copy and paste or stream data to be analyzed and visualized. Mainly used in analytics and data science sector

```python
#Pie chart of India's best selling car companies % values:

import plotly.express as px
import plotly.graph_objects as go
company = ['Maruti', 'Hyundai', 'Tata', 'Kia', 'Mahindra']
market_share = [50, 17.4, 7, 5.8, 5.6] 
colors = ['#ed685f', '#ed9858', '#edbe58', '#d4ed58', '#91ed58']
```

```python
fig = px.pie(names = company, values=market_share, title= 'Indian Car Market')
fig.show()
```

```python
#Bar chart with India covid-19 active cases 
import pandas as pd 
import plotly.express as px 
```

```python
data = pd.read_csv('covid_india.csv')   #https://www.kaggle.com/shivanshsinghal107/covid19-india-dataset
data.head()
```

```python
data.isnull().sum()
```

```python
data2 = data.sort_values(by = ['Active Cases'], ascending= False)
```

```python
data2
```

```python
data3 = data2.head(10)
data3
```

```python
data3 = pd.DataFrame(data3)
```

```python
fig = px.bar(data3, x= 'Name of State / UT', y= 'Active Cases', title = 'Top 10 covid-19 active states in India', color= 'Active Cases')
fig.show( )
```
